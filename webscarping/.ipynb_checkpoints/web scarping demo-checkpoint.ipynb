```python
import requests
from bs4 import BeautifulSoup

url = "https://www.meesho.com/search?q=watch&searchType=POPULAR_SEARCHES&searchIdentifier=text_search"

response = requests.get(url)

response
```

```python
value = BeautifulSoup(response.content,"html")
print(value.prettify())
```

# find

```python
spans = value.find_all("p",class_= "sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU")
spans2 = value.find_all("h5",class_ = "sc-eDvSVe dwCrSh")
for span,span1 in zip(spans,spans2):
    print(span.text.strip())
    print(span1.text.strip())
print("hello")  
```
