```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
num_samples = 500  # Number of rows
data = {
    "Area (sqft)": np.random.randint(800, 4000, num_samples),  # Random area in sqft
    "Bedrooms": np.random.randint(1, 6, num_samples),  # Number of bedrooms
    "Bathrooms": np.random.randint(1, 4, num_samples),  # Number of bathrooms
    "Location Score": np.random.uniform(1, 10, num_samples),  # Location rating
    "Age of House (years)": np.random.randint(0, 50, num_samples),  # Age of the house
    "Nearby Amenities Score": np.random.uniform(1, 10, num_samples),  # Amenity rating
}

# Simulate house prices based on a linear relationship with added noise
weights = [120, 50, 70, 10000, -200, 1500]  # Arbitrary weights for each feature
bias = 50000  # Base price of a house
noise = np.random.randn(num_samples) * 20000  # Add some noise to simulate real-world data
data["Price"] = (
    bias
    + data["Area (sqft)"] * weights[0]
    + data["Bedrooms"] * weights[1]
    + data["Bathrooms"] * weights[2]
    + data["Location Score"] * weights[3]
    + data["Age of House (years)"] * weights[4]
    + data["Nearby Amenities Score"] * weights[5]
    + noise
)

# Convert to a DataFrame
house_data = pd.DataFrame(data)

# Display the first few rows
print(house_data.head())

# Save to a file (optional)
house_data.to_csv("synthetic_house_data.csv", index=False)

# Plot a sample feature vs. price
plt.scatter(house_data["Area (sqft)"], house_data["Price"], alpha=0.5)
plt.title("House Prices vs. Area")
plt.xlabel("Area (sqft)")
plt.ylabel("Price")
plt.show()
```
