## PreProcessing

** Splitting the data to Feature and Label Vector

```python
y=df.target.copy()
x=df.drop('target',axis=1)
```

```python
print(df.columns.tolist())
```

Splitting the Features to ** Train and ** Validation Data

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test  = train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
from sklearn.compose import make_column_selector as mcs
num_sel = mcs(dtype_exclude='object')
cat_sel = mcs(dtype_include='object')
```

** Categorical Feature : One Hot Encoder and ** Numerical Feaeture : Standard Scalar

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler,RobustScaler

cat_pipe = Pipeline([("imputer",SimpleImputer(strategy="most_frequent")),
                          ("encoder",OneHotEncoder(handle_unknown="ignore",
                         sparse_output=False))])

cat_trf=ColumnTransformer([("cat",cat_pipe,[1,2,3,4,6,7,8,13,14])],
                         remainder="passthrough").set_output(transform="pandas")

num_pipe=Pipeline([('imputer',SimpleImputer(strategy="mean")),
                           ('scalar',RobustScaler())])

num_trf=ColumnTransformer([("num",num_pipe,[0,5,9,15,16,17,18])],
                         remainder="passthrough").set_output(transform="pandas")

####trf = ColumnTransformer([("cat",cat_pipe,[1,2,3,4,6,7,8,13,14]),
####                         ("num",num_pipe,[0,5,9,15,16,17])]).set_output(transform='pandas')

pipe = Pipeline([("cat_pipe",cat_trf),
                ("num_pipe",num_trf)])

pipe.fit(X_train)

####trf.fit(X_train)

```

```python
X_train_trf = pipe.transform(X_train)
X_test_trf=pipe.transform(X_test)
```
