```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_youtube_analytics.csv")
df.head()
```

```python
df["dislikes"].unique()
```

```python
df = df.drop(columns = ["video_id","day"])
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
def Handle_zero(n):
    if n < 0:
        return 0
    else:
        return n

for i in df.columns:
    df[i] = df[i].apply(lambda n: Handle_zero(n))
```

```python
df.describe()
```

```python
import pandas as pd
from scipy.stats import zscore

z_scores = df.apply(zscore)

outliers = (z_scores > 3) | (z_scores < -3)

df_cleaned = df[~outliers.any(axis=1)]  

data = df_cleaned
```

```python
data.isnull().sum()
```

```python
# Check for NaN values in the DataFrame
print(data.isna().sum())
```

# so this columns are having nan that means the columns contains 0 only so we are droping it

```python
data.drop(['annotationCloseRate', 'annotationClickThroughRate', 'annotationImpressions', 
           'annotationClickableImpressions', 'annotationClosableImpressions', 'annotationClicks', 
           'annotationCloses', 'cardClickRate', 'cardTeaserClicks', 'subscribersLost', 'cardClicks'], 
          axis=1, inplace=True)
```

```python
import numpy as np  # Use log1p to handle zero values
for i in data.columns:
    data[i] = np.log1p(data[i])
```

```python
data
```

```python
from scipy.stats import skew
for i in data.columns:
    print(i,"   ",skew(data[i]))
```

```python
data["dislikes"].unique()
```

```python
skewness_values = [skew(data[column], nan_policy='omit') for column in data.columns]

# Plot the skewness for each column
plt.figure(figsize=(10, 6))
plt.bar(data.columns, skewness_values, color='skyblue')
plt.xlabel('Features')
plt.ylabel('Skewness')
plt.title('Skewness of Each Feature')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Loop through each column and plot histograms
for i, column in enumerate(data.columns):
    plt.subplot(len(data.columns)//3 + 1, 3, i+1)  # Adjust number of rows and columns for better layout
    plt.hist(data[column].dropna(), bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
