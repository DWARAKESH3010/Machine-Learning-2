```python
df = pd.read_csv("netflix_titles_nov_2019.csv")
df.head()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

# Assuming you already have the DataFrame `df`
df_pca = df.drop(columns=['show_id', 'title', 'description', 'type', 'date_added', 'cast', 'director'])

# Fill NaN values in relevant columns
df_pca = df_pca.fillna('Unknown')

# Extract numerical value from 'duration'
df_pca['duration_num'] = df_pca['duration'].str.extract('(\d+)').astype(float)
df_pca['duration_type'] = df_pca['duration'].str.extract('([a-zA-Z]+)')

# Drop original 'duration' and one-hot encode 'duration_type'
df_pca = df_pca.drop(columns=['duration'])
df_pca_encoded = pd.get_dummies(df_pca, columns=['country', 'listed_in', 'rating', 'duration_type'], drop_first=True)

# Standardize numerical features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_pca_encoded)

# Apply PCA with the desired number of components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Convert PCA result into a DataFrame for easier interpretation
pca_df = pd.DataFrame(data=pca_data, columns=['Principal Component 1', 'Principal Component 2'])

print(pca_df.head())
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Principal Component 1 vs Principal Component 2
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['Principal Component 1'], pca_df['Principal Component 2'], alpha=0.7, edgecolor='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: 2-Component Projection')
plt.show()
```
