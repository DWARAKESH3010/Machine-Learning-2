```python
import pandas as pd
from scipy.stats import zscore

z_scores = df.apply(zscore)

outliers = (z_scores > 3) | (z_scores < -3)

df_cleaned = df[~outliers.any(axis=1)]  

data = df_cleaned
```

```python
import numpy as np

df['log_column'] = np.log1p(df['redViews'])  # log1p is log(x+1), safe for 0 values

# Check the distribution after transformation
sns.histplot(df['log_column'], kde=True)
plt.show()
```
