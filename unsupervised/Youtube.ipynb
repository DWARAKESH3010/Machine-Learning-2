```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_youtube_analytics.csv")
df.head()
```

```python
df["dislikes"].unique()
```

```python
df = df.drop(columns = ["video_id","day"])
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import pandas as pd
from scipy.stats import zscore

z_scores = df.apply(zscore)

outliers = (z_scores > 3) | (z_scores < -3)

df_cleaned = df[~outliers.any(axis=1)]  

data = df_cleaned
```

```python
data.isnull().sum()
```

```python
# Check for NaN values in the DataFrame
print(data.isna().sum())
```

# so this columns are having nan that means the columns contains 0 only so we are droping it

```python
data.drop(['annotationCloseRate', 'annotationClickThroughRate', 'annotationImpressions', 
           'annotationClickableImpressions', 'annotationClosableImpressions', 'annotationClicks', 
           'annotationCloses', 'cardClickRate', 'cardTeaserClicks', 'subscribersLost', 'cardClicks'], 
          axis=1, inplace=True)
```

```python
import numpy as np  # Use log1p to handle zero values
for i in data.columns:
    data[i] = np.log1p(data[i])
```
